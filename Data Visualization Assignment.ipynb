## Problem Statement 1: In this assignment students have to transform iris data into 3 dimensions  and plot a 3d chart with transformed dimensions and colour each data  point with specific class. 

```python
#Importing the libraries
import pandas as pd
from sklearn import decomposition
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
#Importing the dataset Iris
```

```python
df=pd.read_csv('iris.csv')
df.head(6)
```

```python
print('{} \n{}'.format(df.shape, df.variety.value_counts()))
```

```python
#Splitting the data into features and target variable x and y
x=df.iloc[:,0:4]
y=df.iloc[:,4]
x.head()
```

```python
#Encoding the categorical target variable
LabelEncoder()
y=LabelEncoder().fit_transform(y)
y
```

```python
#Transforming our 4 features into 3 using Principal Component Analysis
pca=decomposition.PCA(n_components=3)
x=pca.fit_transform(x)
x #Nested list within an array
```

```python
#Plotting the 3-Dimensional graph and 
fig=plt.figure(1, figsize=(4,3))
ax=Axes3D(fig, rect=[0,0,0.95,1], elev=60, azim=100)
ax.scatter(x[:,0],x[:,1],x[:,2], c=y,edgecolor='k')
plt.show()
```

```python
fig=plt.figure(1, figsize=(4,3))
ax=Axes3D(fig, rect=[0,0,0.95,1], elev=60, azim=130)
ax.scatter(x[:,0],x[:,1],x[:,2], c=y,edgecolor='k')
plt.show()
```
